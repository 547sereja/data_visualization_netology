```python
import pandas as pd
import matplotlib.pyplot as plt
path = '/Users/apple/Desktop/data visualisation'
```

```python
read_data = pd.read_csv(f'{path}/archive/tmdb_5000_movies.csv')
read_data['count'] = 1
```

```python
language_movies = read_data.groupby(['original_language']).sum().sort_values(by='count', ascending=False)
```

```python
language_movies.plot(kind='pie', y='count')
```

### Above we can see obviously domination English language on movies production

```python
most_expensive = read_data.sort_values('budget', ascending=False).head(6)
```

```python
plt.barh(most_expensive['genres'], most_expensive['budget'])
```
